<div class="alert alert-block alert-info">
<b>Naloga: </b>V podanem series obdržite top 2 vrednosti, ki se največkrat pojavita. Ostale vrednosti nadomestite z "other"</div>

```python
np.random.RandomState(100)
ser = pd.Series(np.random.randint(1, 5, [12]))
ser
```

```python
#Rešitev
top_values = ser.value_counts().iloc[0:2].reset_index(level=0)
top_values
condition = (ser == top_values.iloc[0,0]) | (ser == top_values.iloc[1,0])
ser[~condition] = "other"
```
